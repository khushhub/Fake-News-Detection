```python
import pandas as pd
import numpy as np
import re
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

```python
from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
```

```python
from sklearn.metrics import (accuracy_score, classification_report, precision_score,
                             recall_score, f1_score, confusion_matrix, roc_auc_score,
                             roc_curve)
from sklearn.calibration import CalibratedClassifierCV
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline as ImbPipeline
import warnings
warnings.filterwarnings('ignore')
```

1. Download NLTK Resources

```python
nltk.download('stopwords', quiet=True)
nltk.download('punkt', quiet=True)
stop_words = set(stopwords.words('english'))
```

2. Load Dataset with Proper Handling

```python

# Load ISOT datasets
fake = pd.read_csv("Fake.csv")
true = pd.read_csv("True.csv")

fake["label"] = 1
true["label"] = 0

fake = fake[["title", "text", "label"]]
true = true[["title", "text", "label"]]

# Load WELFake dataset
welfake = pd.read_csv("WELFake_Dataset.csv")
welfake = welfake[["title", "text", "label"]]


fake = fake.fillna("")
true = true.fillna("")
welfake = welfake.fillna("")


data = pd.concat([fake, true, welfake], axis=0, ignore_index=True)


data = data.sample(frac=1, random_state=42).reset_index(drop=True)

print("=" * 60)
print("DATASET STATISTICS")
print("=" * 60)
print(f"Total samples: {len(data)}")
print(f"\nLabel distribution:")
print(data["label"].value_counts())
print(f"\nLabel percentages:")
print(data["label"].value_counts(normalize=True) * 100)
print("=" * 60)
```

3. IMPROVED Text Preprocessing

```python
def clean_text_improved(text):
    """
    Improved cleaning that preserves some important signals
    """
    text = str(text).lower()

    url_count = len(re.findall(r'http[s]?://\S+', text))
    text = re.sub(r'http[s]?://\S+', ' urltoken ', text)

    text = re.sub(r'\S+@\S+', ' emailtoken ', text)

    text = re.sub(r'\d+', ' numtoken ', text)

    text = re.sub(r'[^a-z\s]', ' ', text)

    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words and len(word) > 2]

    return " ".join(tokens)


# Apply preprocessing
print("\nPreprocessing text...")
data["title_clean"] = data["title"].apply(clean_text_improved)
data["text_clean"] = data["text"].apply(clean_text_improved)
data["content"] = data["title_clean"] + " " + data["text_clean"]

# Remove empty rows
data = data[data["content"].str.strip() != ""]
data = data.reset_index(drop=True)

print(f"Samples after removing empty content: {len(data)}")
```

4. Feature Engineering

```python
def extract_additional_features(df):
    """
    Extract linguistic and statistical features
    """
    features = pd.DataFrame()

    # Title features
    features['title_length'] = df['title'].str.len()
    features['title_word_count'] = df['title'].str.split().str.len()
    features['title_has_question'] = df['title'].str.contains('\?', regex=True).astype(int)
    features['title_has_exclamation'] = df['title'].str.contains('!', regex=True).astype(int)
    features['title_all_caps_ratio'] = df['title'].apply(
        lambda x: sum(1 for c in str(x) if c.isupper()) / (len(str(x)) + 1)
    )

    # Text features
    features['text_length'] = df['text'].str.len()
    features['text_word_count'] = df['text'].str.split().str.len()
    features['avg_word_length'] = df['text'].apply(
        lambda x: np.mean([len(word) for word in str(x).split()]) if str(x).split() else 0
    )

    # Sentiment indicators (simple)
    features['uppercase_count'] = df['text'].apply(lambda x: sum(1 for c in str(x) if c.isupper()))
    features['exclamation_count'] = df['text'].str.count('!')
    features['question_count'] = df['text'].str.count('\?')

    return features


print("\nExtracting additional features...")
additional_features = extract_additional_features(data)
print(f"Additional features shape: {additional_features.shape}")
```

5. OPTIMIZED TF-IDF


```python
print("\nCreating TF-IDF features ")


word_vectorizer = TfidfVectorizer(
    max_features=3000,
    ngram_range=(1, 2),
    stop_words='english',
    min_df=5,
    max_df=0.85,
    sublinear_tf=True
)

X_word = word_vectorizer.fit_transform(data["content"])
print(f"Word TF-IDF shape: {X_word.shape}")

# Combine features
from scipy.sparse import hstack

X_combined = hstack([X_word, additional_features.values])
print(f"Combined feature shape: {X_combined.shape}")
```

6. Handle Class Imbalance

```python
y = data["label"].values

# Split first
X_train, X_test, y_train, y_test = train_test_split(
    X_combined,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)

print("\n" + "=" * 60)
print("TRAIN-TEST SPLIT")
print("=" * 60)
print(f"Training set: {X_train.shape}")
print(f"Testing set: {X_test.shape}")
print(f"\nTraining label distribution:")
print(pd.Series(y_train).value_counts())
print(f"\nTesting label distribution:")
print(pd.Series(y_test).value_counts())

# Apply SMOTE to training data only
print("\n" + "=" * 60)
print("APPLYING SMOTE FOR BALANCED TRAINING")
print("=" * 60)

smote = SMOTE(random_state=42, k_neighbors=5)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

print(f"Balanced training set: {X_train_balanced.shape}")
print(f"Balanced label distribution:")
print(pd.Series(y_train_balanced).value_counts())
print("=" * 60)
```

7. Model Training

```python
print("\nTraining models ")


models = {
    "Logistic Regression": LogisticRegression(
        max_iter=1000,
        class_weight='balanced',
        C=1.0, #(Regularization parameter)
        random_state=42,
        solver='lbfgs',
        n_jobs=-1
    ),
    "Linear SVM": LinearSVC(
        max_iter=2000,
        class_weight='balanced',
        C=0.1,
        random_state=42,
        dual=False
    )
}

results = {}
trained_models = {}
calibrated_models = {}

for name, model in models.items():
    print(f"\n{'=' * 70}")
    print(f"Training: {name}")
    print(f"{'=' * 70}")

    # Train on balanced data
    model.fit(X_train_balanced, y_train_balanced)
    trained_models[name] = model

    # OPTIMIZATION: Calibration with fewer folds
    print(f"  Calibrating {name}...")
    calibrated = CalibratedClassifierCV(model, cv=3, method='sigmoid', n_jobs=-1)
    calibrated.fit(X_train_balanced, y_train_balanced)
    calibrated_models[name] = calibrated

    # Predict on test set
    print(f"  Predicting on test set...")
    y_pred = calibrated.predict(X_test) #Predicts class labels (0 or 1)

    # Get probabilities
    y_pred_proba = calibrated.predict_proba(X_test)[:, 1]

    # Calculate metrics
    acc = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, zero_division=0)
    recall = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    auc = roc_auc_score(y_test, y_pred_proba)

    results[name] = {
        "Accuracy": acc,
        "Precision": precision,
        "Recall": recall,
        "F1-score": f1,
        "AUC-ROC": auc
    }

    print(f"\n  Results:")
    print(f"  Accuracy:  {acc:.4f}")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall:    {recall:.4f}")
    print(f"  F1-score:  {f1:.4f}")
    print(f"  AUC-ROC:   {auc:.4f}")
```

8. Model Comparison

```python
print("\n" + "=" * 70)
print("MODEL COMPARISON")
print("=" * 70)
results_df = pd.DataFrame(results).T
print(results_df.to_string())
print("=" * 70)

# Select best model based on F1-score
best_model_name = results_df['F1-score'].idxmax()
best_model = calibrated_models[best_model_name]
print(f"\n✓ Best Model Selected: {best_model_name}")
print(f"  F1-Score: {results_df.loc[best_model_name, 'F1-score']:.4f}")
```

9. Generate Confusion Matrices

```python
print("\nGenerating visualizations...")

for name in models.keys():
    y_pred = calibrated_models[name].predict(X_test)
    cm = confusion_matrix(y_test, y_pred)

    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
                xticklabels=['Real', 'Fake'],
                yticklabels=['Real', 'Fake'])
    plt.title(f"Confusion Matrix - {name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()
```

11. Enhanced Prediction Function

```python
def predict_news(text, title="", threshold=0.5):
    """
    Fast prediction with confidence scores (self-contained version)
    """
    # Import needed here to ensure availability
    import re
    import pandas as pd
    import numpy as np
    from scipy.sparse import hstack
    from nltk.corpus import stopwords

    # Get stopwords INSIDE the function
    stop_words_local = set(stopwords.words('english'))

    # Helper function for cleaning (embedded inside)
    def clean_text_local(txt):
        txt = str(txt).lower()
        txt = re.sub(r'http[s]?://\S+', ' urltoken ', txt)
        txt = re.sub(r'\S+@\S+', ' emailtoken ', txt)
        txt = re.sub(r'\d+', ' numtoken ', txt)
        txt = re.sub(r'[^a-z\s]', ' ', txt)
        txt = re.sub(r'\s+', ' ', txt).strip()
        tokens = [w for w in txt.split() if w not in stop_words_local and len(w) > 2]
        return " ".join(tokens)

    # Helper function for features - MUST MATCH TRAINING FEATURES
    def extract_features_local(df):
        features = pd.DataFrame()

        # Title features
        features['title_length'] = df['title'].str.len()
        features['title_word_count'] = df['title'].str.split().str.len()
        features['title_has_question'] = df['title'].str.contains('\?', regex=True).astype(int)
        features['title_has_exclamation'] = df['title'].str.contains('!', regex=True).astype(int)
        features['title_all_caps_ratio'] = df['title'].apply(
            lambda x: sum(1 for c in str(x) if c.isupper()) / (len(str(x)) + 1)
        )

        # Text features
        features['text_length'] = df['text'].str.len()
        features['text_word_count'] = df['text'].str.split().str.len()
        features['avg_word_length'] = df['text'].apply(
            lambda x: np.mean([len(word) for word in str(x).split()]) if str(x).split() else 0
        )

        # Sentiment indicators
        features['uppercase_count'] = df['text'].apply(lambda x: sum(1 for c in str(x) if c.isupper()))
        features['exclamation_count'] = df['text'].str.count('!')
        features['question_count'] = df['text'].str.count('\?')

        return features

    # Clean text
    title_clean = clean_text_local(title)
    text_clean = clean_text_local(text)
    content = title_clean + " " + text_clean

    # Extract features
    temp_df = pd.DataFrame({'title': [title], 'text': [text]})
    add_features = extract_features_local(temp_df)

    # Transform
    word_vec = word_vectorizer.transform([content])
    combined_vec = hstack([word_vec, add_features.values])

    # Predict
    proba = best_model.predict_proba(combined_vec)[0]
    fake_prob = proba[1]
    real_prob = proba[0]

    final_prediction = 1 if fake_prob >= threshold else 0

    return {
        "prediction": "FAKE NEWS" if final_prediction == 1 else "REAL NEWS",
        "fake_probability": fake_prob,
        "real_probability": real_prob,
        "confidence": max(fake_prob, real_prob)
    }


print("\n✓ Prediction function ready!")
```
